```python
import pandas as pd
import psycopg2
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
# Connect to my database
# Database connection parameters
conn = psycopg2.connect(
    host="localhost",
    port="4228",  
    database="postgres",  
    user="postgres",  
    password="Blairp11"  
)

cursor = conn.cursor()
```

```python
# Query for UCLA offensive stats along with weather, altitude, and distance
query = """
SELECT date, raw_location, opponent, 
       passing_cmp, passing_att, completion_pct, passing_yds, passing_td, 
       rushing_att, rushing_yds, rushing_avg, rushing_td, total_plays, total_yards, avg_yds,
       pass_first_downs, rush_first_downs, first_down_pens, total_first_downs, 
       total_penalties, total_penalties_yds, total_fum, total_int, total_to,
       temp_f, precipitation_in, windspeed_mph, altitude_m, distance_from
FROM UCLA_Offensive_Stats;
"""

ucla_data = pd.read_sql(query, conn)


ucla_data.head()
```

```python
from sklearn.linear_model import LinearRegression

X_wind = ucla_data['windspeed_mph'].values.reshape(-1, 1)
y_wind = ucla_data['passing_yds'].values


combined_wind = pd.concat([ucla_data['windspeed_mph'], ucla_data['passing_yds']], axis=1).dropna()
X_wind_clean = combined_wind['windspeed_mph'].values.reshape(-1, 1)
y_wind_clean = combined_wind['passing_yds'].values

# Fit model
model_wind = LinearRegression()
model_wind.fit(X_wind_clean, y_wind_clean)


y_wind_pred = model_wind.predict(X_wind_clean)


plt.figure(figsize=(8, 6))
plt.scatter(X_wind_clean, y_wind_clean, color='blue', label='Actual Data')
plt.plot(X_wind_clean, y_wind_pred, color='red', linewidth=2, label='Linear Fit')
plt.xlabel('Windspeed (mph)')
plt.ylabel('Passing Yards')
plt.title('Passing Yards vs Windspeed')
plt.legend()
plt.show()
```

```python
# windspeed ranges
ucla_data['Windspeed_Range'] = pd.cut(ucla_data['windspeed_mph'], bins=[0, 5, 10, 15, 20, 25, 30], 
                                      labels=['0-5 mph', '6-10 mph', '11-15 mph', '16-20 mph', '21-25 mph', '26-30 mph'])


wind_grouped = ucla_data.groupby('Windspeed_Range').agg({
    'passing_yds': 'mean',
    'rushing_yds': 'mean',
    'total_yards': 'mean',
    'completion_pct': 'mean',
    'passing_td': 'mean',
    'rushing_att': 'mean',
    'rushing_td': 'mean',
    'total_first_downs': 'mean',
    'total_to': 'mean'
}).reset_index()


fig, axs = plt.subplots(3, 3, figsize=(18, 12))

# Bar plots for different metrics
sns.barplot(x='Windspeed_Range', y='passing_yds', data=wind_grouped, palette='coolwarm', ax=axs[0, 0])
axs[0, 0].set_title('Passing Yards vs Windspeed')

sns.barplot(x='Windspeed_Range', y='rushing_yds', data=wind_grouped, palette='coolwarm', ax=axs[0, 1])
axs[0, 1].set_title('Rushing Yards vs Windspeed')

sns.barplot(x='Windspeed_Range', y='total_yards', data=wind_grouped, palette='coolwarm', ax=axs[0, 2])
axs[0, 2].set_title('Total Yards vs Windspeed')

sns.barplot(x='Windspeed_Range', y='completion_pct', data=wind_grouped, palette='coolwarm', ax=axs[1, 0])
axs[1, 0].set_title('Completion Percentage vs Windspeed')

sns.barplot(x='Windspeed_Range', y='passing_td', data=wind_grouped, palette='coolwarm', ax=axs[1, 1])
axs[1, 1].set_title('Passing Touchdowns vs Windspeed')

sns.barplot(x='Windspeed_Range', y='rushing_att', data=wind_grouped, palette='coolwarm', ax=axs[1, 2])
axs[1, 2].set_title('Rushing Attempts vs Windspeed')

sns.barplot(x='Windspeed_Range', y='rushing_td', data=wind_grouped, palette='coolwarm', ax=axs[2, 0])
axs[2, 0].set_title('Rushing Touchdowns vs Windspeed')


```
